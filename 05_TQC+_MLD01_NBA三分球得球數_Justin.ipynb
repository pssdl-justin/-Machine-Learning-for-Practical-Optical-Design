```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
```

```
NBApoints_data= pd.read_csv("NBApoints.csv")
print(NBApoints_data.shape)
```

**資料說明**
* Pk：排名
* Player：球員
* Pos：守備位置
* Age：年齡
* Tm：隊名

```
#TODO
from sklearn.preprocessing import LabelEncoder
label_encoder_conver = LabelEncoder()
NBApoints_data["Pos"] = label_encoder_conver.fit_transform(NBApoints_data["Pos"])

label_encoder_conver = LabelEncoder()
NBApoints_data["Tm"] = label_encoder_conver.fit_transform(NBApoints_data["Tm"])
```

```
print(NBApoints_data.shape)
display(NBApoints_data)
```

```
X_train = pd.DataFrame(NBApoints_data.T, ['Pos', 'Age', 'Tm']).T
y_train = pd.DataFrame(NBApoints_data.T, ['3P']).T
y_train = np.array(y_train, dtype='float')
print(X_train.shape, y_train.shape)
```

```
X = NBApoints_data.iloc[:, 2:5].values
print(X.shape)
print(X)
```

```
#TODO
NBApoints_linear_model = LinearRegression()
NBApoints_linear_model.fit(X_train, y_train)
```

```
NBApoints_linear_model_predict_result = NBApoints_linear_model.predict([[5, 28, 10]])
r_squared = NBApoints_linear_model.score(X_train, y_train)

print("三分球得球數=",NBApoints_linear_model_predict_result)
print("R_squared值=",r_squared)
```

```
f_regressor = f_regression
# print("F score and p-value：", f_regressor(X_train, NBApoints_data["3P"]))
print("F score and p-value：", f_regressor(X_train, y_train))
```

# statsmodels

```
import statsmodels.api as sm
X_train_c = sm.add_constant(X_train)
X_train_c = np.array(X_train_c, dtype='float')

regressor_OLS = sm.OLS(y_train, X_train_c).fit()
print(regressor_OLS.summary())
```

```
X_opt = np.array(X_train_c[:, [0, 1, 2]], dtype='float')

regressor_OLS = sm.OLS(y_train, X_opt).fit()
print(regressor_OLS.summary())
```

```
X_opt = np.array(X_train_c[:, [0, 1]], dtype='float')

regressor_OLS = sm.OLS(y_train, X_opt).fit()
print(regressor_OLS.summary())
```
